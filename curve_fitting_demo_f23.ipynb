```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as fit
```

```python
m = 0.5
b = 3#arbitrary parameters for a linear dataset
sigma = 1 # noise
n = 20 # number of points to generate
xdata = np.arange(0,n)
ydata = m *xdata + b + sigma*np.random.standard_normal(size=n) # generate data with spread
```

```python
# check data points by visualizing in a graph
plt.figure(figsize=(8,6))
plt.scatter(xdata, ydata)
plt.show()
```

```python
def f(x, m, b):
    return m * x + b
# y = mx+b
# y = ax^2 + bx + c

p0 = [0.5, 2]

parameters, covariance_matrix = fit.curve_fit(f, xdata, ydata, p0)
```

```python
m = parameters[0]
b = parameters[1]
m_error = (covariance_matrix[0,0]) **0.5
b_error = (covariance_matrix[1,1]) ** 0.5
```

```python
print('the parameters are: ', parameters)
print('m = {0:1.5f} +\- {1:1.5f}'.format(m,m_error))
```
