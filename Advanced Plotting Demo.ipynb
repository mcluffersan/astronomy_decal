```python
#import statements
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x,y = np.meshgrid(np.linspace(-1,1,100), np.linspace(2,-2,100))
f = x + y

plt.figure()
plt.imshow(f, cmap = 'inferno', origin = 'lower')
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-1,1,100), np.linspace(2,-2,100))
a = np.sqrt(x**2+y**2)

plt.figure()
plt.imshow(a)
plt.show()
```
